```python
import pandas as pd 
import numpy as np 
import  matplotlib.pyplot as plt

import seaborn as sns
```

```python
#Generate descriptive statistics that summarize the central tendency
df.describe(include="all")
```

```python
#Plot pairwise relationships in a dataset.
sns.pairplot(df) 
```

```python
#input features and target variables

Radio=df.iloc[:,2:3]
Sales=df.iloc[:,4:5].values
```

```python
#normally distribute data use sklearn MinMaxScaler to scale the data
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()


Radio = sc.fit_transform(Radio)

#(is a 1 dimensional columnar array
Sales = Sales.reshape(-1,1)

Sales = sc.fit_transform(Sales)
```

```python
# Split arrays or matrices into random train and test subsets
#test data 0.3 and remaining 0.7 data will be train

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Radio, Sales, test_size=0.3)
```

####  Returns the Keras sequential model. with 13 nodes single layer 

```python
from keras import Sequential
from keras.layers import Dense
def build_regressor():
    # Sequential() is Linear stack of layers
    regressor = Sequential()
    regressor.add(Dense(13, input_shape=(1,)))
    regressor.add(Dense(units=1))
    regressor.compile(optimizer='adam', loss='mean_squared_error',  metrics=['mae','accuracy'])
    return regressor
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
regressor = KerasRegressor(build_fn=build_regressor, batch_size=32,epochs=100)
```

```python
results=regressor.fit(X_train,y_train)
```

```python
x = np.array([2.1])


y_pred= regressor.predict(x)
y_pred.
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
